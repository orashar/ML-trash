```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line plot

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x, y1, y2)
```

```python
themes = plt.style.available
plt.style.use('seaborn')
```

```python
plt.plot(x, y1, color='red', label="apple", marker='o')
plt.plot(x, y2, color='green', label="kiwi", linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title("Price of Fruitsover Time")
plt.legend()
plt.show()
```

```python
arr = np.array([1, 2, 3, 4])**2
plt.plot(arr)
plt.show()
```

# Scatter plot

```python
# adjust size of plot
plt.figure(figsize=(5,5))

plt.scatter(x, y1, color='red', label="apple", marker='^')
plt.scatter(x, y2, color='green', label="kiwi", linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title("Price of Fruitsover Time")
plt.legend()
plt.show()
```

# Bar Graphs

```python
plt.style.use('dark_background')

x = np.array([0,1,2])*2

plt.bar(x-0.25, [10,20,15], width=0.5, label='current year', tick_label=['Gold', 'Platinum', 'Silver'])
plt.bar(x+0.25, [20,10,12], width=0.5, label='next year')
plt.xlabel('Metal')
plt.ylabel('price')
plt.ylim(0, 40)
plt.xlim(-2, 5)
plt.legend()
plt.show()
```

#  PIE chart

```python
plt.style.use("seaborn")
```

```python
subjects ="Maths", "Chem", "Pyhsics", "English"
weightage = [20, 10, 15, 5]

plt.pie(weightage, labels=subjects, explode=(0,0,0.1,0), autopct="%1.1f%%", shadow=True)
plt.show()
```

# Data from a Standard Normalised Data

```python
xsn = np.random.randn(100)

u = 60
sigma = 5

x = np.round(xsn*sigma + u)
x2 =np.round(xsn*15 +40)
x
x2
```

# Histogram

```python
plt.hist(x, alpha=0.8, label="Maths")
plt.hist(x2, alpha=0.8, label="Physics")
plt.ylabel("Frequency of Students")
plt.xlabel("Marks range")
plt.legend()
plt.show()
```

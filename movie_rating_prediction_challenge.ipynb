```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import pandas as pd
```

```python
tokenizer = RegexpTokenizer(r'\w+')
en_stopwords = set(stopwords.words("english"))
ps = PorterStemmer()
cv = CountVectorizer(ngram_range=(1,2))
```

```python
df = pd.read_csv("./Train/Train.csv")
dft = pd.read_csv("./Test/Test.csv")

X = df.values[:,0]
Y = df.values[:,1]

Xt = dft.values.reshape(-1,)

print(X.shape, Y.shape, Xt.shape)
```

```python
def getCleanedReview(review):
    review = review.lower()
    review = review.replace("<br></br>", " ")
    
    #tokenize
    tokens = tokenizer.tokenize(review)
    new_tokens = (token for token in tokens if token not in en_stopwords)
    
    stemmed_tokens = (ps.stem(token) for token in new_tokens)
    
    cleaned_review = " ".join(stemmed_tokens)
    
    return cleaned_review
```

```python
print("not" in en_stopwords)
en_stopwords.remove("not")
print("not" in en_stopwords)
```

```python
Xc = [ getCleanedReview(x) for x in X]
Xtc = [ getCleanedReview(x) for x in Xt]
```

```python
print(Xc[0], Xtc[0])
```

```python
Xv = cv.fit_transform(Xc)
```

```python
print(Xv.shape, Xv[0])
cv.get_feature_names()
```

```python
Xtv = cv.transform(Xtc)
```

```python
print(Xtv.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(Xv, Y)
```

```python
Yt = mnb.predict(Xtv)
```

```python
mnb.score(Xv, Y)
```

```python
dfyt = pd.DataFrame(Yt)
print(dfyt.shape)
print(dfyt.head())
```

```python
dfyt.to_csv("./predicted_output_movie.csv", index=True)
```

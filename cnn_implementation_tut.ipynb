```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImg(img, title="Image"):
    plt.imshow(img, cmap="gray")
    plt.axis('off')
    plt.title(title + str(img.shape))
    plt.style.use('seaborn')
    plt.show()
```

```python
img = cv2.imread("./cody2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (100,100))
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
drawImg(img)
drawImg(img_gray)
```

```python
def convolution(img, img_filter):
    W = img.shape[0]
    H = img.shape[1]
    
    F = img_filter.shape[0]
    
    new_img = np.zeros((W-F+1, H-F+1))
    
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    new_img[row][col] += img[row+i][col+j]*img_filter[i][j]
                    
                if new_img[row][col] > 255:
                    new_img[row][col] = 255
                    
                elif new_img[row][col] < 0:
                    new_img[row][col] = 0
                    
    return new_img
```

```python
blur_filter = np.ones((3,3))/9.0
edge_filter = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])

output1 = convolution(img_gray, blur_filter)
output2 = convolution(img_gray, edge_filter)
```

```python
drawImg(output1)
drawImg(output2)
drawImg(img_gray)
```

```python
X = np.array([[1,0,2,3],[4,6,6,8],[3,1,1,0],[1,2,2,4]])

def pooling(X, mode='max'):
    stride = 2
    f = 2
    H,W = X.shape
    
    HO = int((H-f)/stride) + 1
    WO = int((W-f)/stride) + 1
    
    output =np.zeros((HO,WO))
    
    for r in range(HO):
        for c in range(WO):
            r_start = r*stride
            c_start = c*stride
            r_end = r_start + f
            c_end = c_start + f
            
            X_slice = X[r_start:r_end, c_start:c_end]
            if mode == 'max':
                output[r][c] = np.max(X_slice)
            else:
                output[r][c] = np.mean(X_slice)
                
    return output
```

```python
pooling_output = pooling(X)
print(pooling_output)
```
